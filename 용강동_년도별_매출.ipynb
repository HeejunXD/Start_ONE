```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('마포구_동별_연평균매출액2021.csv',encoding='cp949')
```

```python
df.set_index('분류',inplace=True)
```

```python
df=df[['용강동']]
df
```

```python
n=0
for year in range(2014,2022):
    if n==0:
        df=pd.read_csv(f'마포구_동별_연평균매출액{year}.csv',encoding='cp949')
        df.set_index('분류',inplace=True)
        df=df[['용강동']]
        df.rename(columns={'용강동':year},inplace=True)
        n=1
    else:
        df_sub=pd.read_csv(f'마포구_동별_연평균매출액{year}.csv',encoding='cp949')
        df_sub.set_index('분류',inplace=True)
        df_sub=df_sub[['용강동']]
        df_sub.rename(columns={'용강동':year},inplace=True)
        df=pd.concat([df,df_sub],axis=1)
```
