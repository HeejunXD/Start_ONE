```python
def save_csv(name, list_code):
  
  global df_kor 
  global df_ch 
  global df_jap 
  global df_usa 
  global df_bake 
  global df_fast
  global df_chicken 
  global df_snack 
  global df_alcoho
  global df_coffee

  df_kor = pd.DataFrame()
  df_ch = pd.DataFrame()
  df_jap = pd.DataFrame()
  df_usa = pd.DataFrame()
  df_bake = pd.DataFrame()
  df_fast = pd.DataFrame()
  df_chicken = pd.DataFrame()
  df_snack = pd.DataFrame()
  df_alcoho = pd.DataFrame()
  df_coffee = pd.DataFrame()
  df_sample3 = ['CS100001', 'CS100002', 'CS100003', 'CS100004', 'CS100005', 'CS100006','CS100007', 'CS100008', 'CS100009', 'CS100010']

  for i in list_code:
    df_kor = pd.concat([df_kor, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[0])]])
    df_ch = pd.concat([df_ch, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[1])]])
    df_jap = pd.concat([df_jap, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[2])]])
    df_usa = pd.concat([df_usa , market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[3])]])
    df_bake = pd.concat([df_bake , market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[4])]])
    df_fast = pd.concat([df_fast, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[5])]])
    df_chicken = pd.concat([df_chicken, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[6])]])
    df_snack = pd.concat([df_snack, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[7])]])
    df_alcoho = pd.concat([df_alcoho, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[8])]])
    df_coffee = pd.concat([df_coffee, market[(market['상권_코드']==i) & (market['서비스_업종_코드']==df_sample3[9])]])

  '''
  df_kor.to_csv(f'{name}_한식')
  df_ch.to_csv(f'{name}_중식')
  df_jap.to_csv(f'{name}_일식')
  df_usa.to_csv(f'{name}_양식')
  df_bake.to_csv(f'{name}_베이커리')
  df_fast.to_csv(f'{name}_패스트푸드')
  df_chicken.to_csv(f'{name}_치킨')
  df_snack.to_csv(f'{name}_분식')
  df_alcoho.to_csv(f'{name}_주점')
  df_coffee.to_csv(f'{name}_커피')'''

  return df_kor, df_ch, df_jap, df_usa , df_bake, df_fast, df_chicken, df_snack, df_alcoho, df_coffee

def age_num(df):
  a = round(df['연령대_10_매출_건수'].sum() / df['점포수'].sum(), 0)
  b = round(df['연령대_20_매출_건수'].sum() / df['점포수'].sum(), 0)
  c = round(df['연령대_30_매출_건수'].sum() / df['점포수'].sum(), 0)
  d = round(df['연령대_40_매출_건수'].sum() / df['점포수'].sum(), 0)
  e = round(df['연령대_50_매출_건수'].sum() / df['점포수'].sum(), 0)
  f = round(df['연령대_60_이상_매출_건수'].sum() / df['점포수'].sum(), 0)

  return a, b, c, d, e, f

def age_won(df):
  a = round(df['연령대_10_매출_금액'].sum() / df['점포수'].sum(), 0)
  b = round(df['연령대_20_매출_금액'].sum() / df['점포수'].sum(), 0)
  c = round(df['연령대_30_매출_금액'].sum() / df['점포수'].sum(), 0)
  d = round(df['연령대_40_매출_금액'].sum() / df['점포수'].sum(), 0)
  e = round(df['연령대_50_매출_금액'].sum() / df['점포수'].sum(), 0)
  f = round(df['연령대_60_이상_매출_금액'].sum() / df['점포수'].sum(), 0)

  return a, b, c, d, e, f

def weekly_won(df):
  mon = round(df['월요일_매출_금액'].sum() / df['점포수'].sum(), -1)
  tue = round(df['화요일_매출_금액'].sum() / df['점포수'].sum(), -1)
  wed = round(df['수요일_매출_금액'].sum() / df['점포수'].sum(), -1)
  thu = round(df['목요일_매출_금액'].sum() / df['점포수'].sum(), -1)
  fri = round(df['금요일_매출_금액'].sum() / df['점포수'].sum(), -1)
  sat = round(df['토요일_매출_금액'].sum() / df['점포수'].sum(), -1)
  sun = round(df['일요일_매출_금액'].sum() / df['점포수'].sum(), -1)

  return mon, tue, wed, thu, fri, sat, sun

def weekly_num(df):
  mon = round(df['월요일_매출_건수'].sum() / df['점포수'].sum(), 0)
  tue = round(df['화요일_매출_건수'].sum() / df['점포수'].sum(), 0)
  wed = round(df['수요일_매출_건수'].sum() / df['점포수'].sum(), 0)
  thu = round(df['목요일_매출_건수'].sum() / df['점포수'].sum(), 0)
  fri = round(df['금요일_매출_건수'].sum() / df['점포수'].sum(), 0)
  sat = round(df['토요일_매출_건수'].sum() / df['점포수'].sum(), 0)
  sun = round(df['일요일_매출_건수'].sum() / df['점포수'].sum(), 0)

  return mon, tue, wed, thu, fri, sat, sun

def time_won(df):
  a = round(df['시간대_00~06_매출_금액'].sum() / df['점포수'].sum(), -1)
  b = round(df['시간대_06~11_매출_금액'].sum() / df['점포수'].sum(), -1)
  c = round(df['시간대_11~14_매출_금액'].sum() / df['점포수'].sum(), -1)
  d = round(df['시간대_14~17_매출_금액'].sum() / df['점포수'].sum(), -1)
  e = round(df['시간대_17~21_매출_금액'].sum() / df['점포수'].sum(), -1)
  f = round(df['시간대_21~24_매출_금액'].sum() / df['점포수'].sum(), -1)

  return a, b, c, d, e, f

def time_num(df):
  a = round(df['시간대_건수~06_매출_건수'].sum() / df['점포수'].sum(), 0)
  b = round(df['시간대_건수~11_매출_건수'].sum() / df['점포수'].sum(), 0)
  c = round(df['시간대_건수~14_매출_건수'].sum() / df['점포수'].sum(), 0)
  d = round(df['시간대_건수~17_매출_건수'].sum() / df['점포수'].sum(), 0)
  e = round(df['시간대_건수~21_매출_건수'].sum() / df['점포수'].sum(), 0)
  f = round(df['시간대_건수~24_매출_건수'].sum() / df['점포수'].sum(), 0)

  return a, b, c, d, e, f

def gender(df):
  man = round(df['남성_매출_비율'].sum() / len(df['남성_매출_비율']), 1)
  woman = round(100 - man, 1)

  return man, woman

def week_weekend(df):
  a = round(df['주중_매출_비율'].sum() / len(df['주중_매출_비율']), 1)
  b = round(100 - a, 1)

  return a, b

'''
def gather():
  a = ['한식', '중식', '일식', '양식', '베이커리', '패스트푸드', '치킨', '분식', '주점', '카페']
  b = [df_kor, df_ch, df_jap, df_usa , df_bake, df_fast, df_chicken, df_snack, df_alcoho, df_coffee]
  for i in range(len(a)):
    print(f'{a[i]}')
    print('연령대 별 매출 건수')
    print(age_num(b[i]))
    print('연령대 별 매출 금액')
    print(age_won(b[i]))
    print('요일 별 매출 건수')
    print(weekly_num(b[i]))
    print('요일 별 매출 금액')
    print(weekly_won(b[i]))
    print('시간대 별 매출 건수')
    print(time_num(b[i]))
    print('시간대 별 매출 금액')
    print(time_won(b[i]))
    print('성비')
    print(gender(b[i]))
    print('주중주말비율')
    print(week_weekend(b[i]))
    print('-------------------------------------------------------------------------------')'''
  
def gather_df (name) :

  df_age_num = pd.DataFrame(index=['10대', '20대', '30대', '40대', '50대', '60대 이상'])
  df_age_won = pd.DataFrame(index=['10대', '20대', '30대', '40대', '50대', '60대 이상'])
  df_weekly_num = pd.DataFrame(index=['월', '화', '수', '목', '금', '토', '일'])
  df_weekly_won = pd.DataFrame(index=['월', '화', '수', '목', '금', '토', '일'])
  df_time_num = pd.DataFrame(index=['~06', '~11', '~14', '~17', '~21', '~24'])
  df_time_won = pd.DataFrame(index=['~06', '~11', '~14', '~17', '~21', '~24'])
  df_gender = pd.DataFrame(index=['남', '여'])
  df_week_weekend = pd.DataFrame(index=['주중', '주말'])


  a = ['한식', '중식', '일식', '양식', '베이커리', '패스트푸드', '치킨', '분식', '주점', '카페']
  b = [df_kor, df_ch, df_jap, df_usa , df_bake, df_fast, df_chicken, df_snack, df_alcoho, df_coffee]

  for i in range(len(a)):
    df_age_num[a[i]] = age_num(b[i])
    df_age_won[a[i]] = age_won(b[i])
    df_weekly_num[a[i]] = weekly_num(b[i])
    df_weekly_won[a[i]] = weekly_won(b[i])
    df_time_num[a[i]] = time_num(b[i])
    df_time_won[a[i]] = time_won(b[i])
    df_gender[a[i]] = gender(b[i])
    df_week_weekend[a[i]] = week_weekend(b[i])

  df_age_num.to_csv(f'{name}_연령대 별 매출 건수')
  df_age_won.to_csv(f'{name}_연령대 별 매출 금액')
  df_weekly_num.to_csv(f'{name}_요일 별 매출 건수')
  df_weekly_won.to_csv(f'{name}_요일 별 매출 금액')
  df_time_num.to_csv(f'{name}_시간대 별 매출 건수')
  df_time_won.to_csv(f'{name}_시간대 별 매출 금액')
  df_gender.to_csv(f'{name}_성별 매출 비율')
  df_week_weekend.to_csv(f'{name}_주중 주말 매출 비율')

  return df_age_num, df_age_won,df_weekly_num , df_weekly_won , df_time_num , df_time_won , df_gender, df_week_weekend
```
